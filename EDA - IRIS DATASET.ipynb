# EDA on IRIS DATASET


```python
import pandas as pd
import seaborn as sn
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the iris dataset (csv file)

iris = pd.read_csv("dataset/iris.csv")
```

```python
# data points and features
print(iris.shape)

# column name in dataset
print(iris.columns)
```

```python
# total number of rows for each class values

iris["Species"].value_counts()
```

# 2D plotting


```python
iris.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
plt.show()
```

```python
#with color coding for 3 different species

sn.set_style('whitegrid')
sn.FacetGrid(iris,hue='Species',size=4).map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()

plt.show()
```

# Pair Plot

```python
plt.close()
sn.set_style('whitegrid')
sn.pairplot(iris,hue='Species',size=3)
plt.show()

# Limitation of this is that if we have n features, we can have many combination
```

# Histogram ( PDF and CDF)

```python
# count number of point per x value
#PDF -> smooth histogram
sn.FacetGrid(iris,hue='Species',height=10).map(sn.histplot,'SepalLengthCm').add_legend()

plt.show()
```

```python
# mean median and variance
# 1. outliner - > data point which is not wrong
# 2. means -> avr behaviour
print('mean')
print(np.mean(iris['SepalLengthCm']))

# 3. spread -> how far the data is spreaded from the mean - > std - devaiation
# 4. root of variance
print('std-dev')
print(np.std(iris['SepalLengthCm']))
```

```python
# Box plot with whiskers
# median percentile and quantiles
# shows box in quantiles (25,50,75) 
# whisker -> min and max value
sn.boxplot(x='Species',y='PetalLengthCm',data=iris)
```

```python
# violin plot
# pdf as well as cdf showing
sn.violinplot(x='Species',y='PetalLengthCm',data=iris,size=8)
```

```python
# multivariate analysis

sn.jointplot(x='PetalLengthCm',y='SepalLengthCm',data=iris,kind='kde')
iris
```
